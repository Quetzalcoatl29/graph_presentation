```python
import os
import sys
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


files = []
basePath = '../graph_generate/results'

def import_file(filename):
    file = basePath + '/' + filename

    data_import = {}
    with open(file) as json_file:
        data_import = json.load(json_file)

    return data_import


for f in os.listdir(basePath):
    if os.path.isfile(os.path.join(basePath, f)) and f.endswith(".json"):
        files.append(f)
```

```python
data_all = {}

data_presentation_1 = {}
for filename in files:
    data_import = import_file(filename)
    
    name = filename.replace(basePath, "").replace(".json", "")
    
    algorithms_presentation = {}
    for data_each_one in data_import:
        algorithms_presentation[data_each_one['algorithm']] = data_each_one['time']
        
    data_all[name] = pd.DataFrame(data_import)
    
data_all
```

```python
for attr, value in data_all.items():
    

    value.plot.bar(x='algorithm', y='time');
```
